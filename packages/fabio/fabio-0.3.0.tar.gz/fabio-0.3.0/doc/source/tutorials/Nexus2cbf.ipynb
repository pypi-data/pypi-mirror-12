# Nexus -> CBF #
In this tutorial we will see how to export a Nexus archive produced by the Eiger detector from Dectris into a bunch of CBF files, similar to the one generated by Pilatus detectors, using the FabIO library to write images and h5py to read them.

**Nota:** HDF5 files produced by Nexus detector use a specific LZ4 plugin for reading/writing. They require recent version of hdf5 (1.8.12 at least).



```python
import fabio, h5py
```

```python
h5_file = h5py.File("series_23_master.h5")
print(list(h5_file.keys()))
```

```python
print(list(h5_file["entry"].keys()))
```

```python
print(h5_file["entry/data_000001"].shape)
```

```python
h5_file["entry/instrument/detector"].keys()
```

```python
dset = h5_file["entry/data_000001"]
for idx, frame in enumerate(dset):
    cbf = fabio.cbfimage.cbfimage(data=frame)
    fname = "series_23_%04i.cbf"%idx
    cbf.write(fname)
    print(fname)
    
    
```

##Conclusion##
FabIO offers a versatile way for manipulating image files and convert them.
